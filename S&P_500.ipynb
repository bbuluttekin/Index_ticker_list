```python
import requests
import urllib2
from pattern import web
import bs4 as bs
import pandas as pd
import numpy as np
```

```python
url = 'http://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
```

```python
snp = requests.get(url)
```

```python
tree = web.Element(snp.text)
```

```python
#for i in range(605):
#    print tree('div.mw-content-ltr')[0].by_tag('tr')[i].content
```

```python
header = []
header.append(tree('div.mw-content-ltr')[0].by_tag('tr')[0].by_tag('a')[0].content)
header.append(tree('div.mw-content-ltr')[0].by_tag('tr')[0].by_tag('th')[1].content)
header.append(tree('div.mw-content-ltr')[0].by_tag('tr')[0].by_tag('a')[3].content)
header.append(tree('div.mw-content-ltr')[0].by_tag('tr')[0].by_tag('th')[4].content)
header.append(tree('div.mw-content-ltr')[0].by_tag('tr')[0].by_tag('th')[5].content)
header.append(tree('div.mw-content-ltr')[0].by_tag('tr')[0].by_tag('th')[6].content)
#header.append(tree('CIK')
header.append('CIK')
header
```

```python
ts = []
sec = []
sector = []
subi = []
add = []
dateadded = []
cik = []
for i in range(1, 503):
    line = tree('div.mw-content-ltr')[0].by_tag('tr')[i]
    ts.append(line.by_tag('a')[0].content)
    sec.append(line.by_tag('a')[1].content)
    sector.append(line.by_tag('td')[3].content)
    subi.append(line.by_tag('td')[4].content)
    #add.append(line.by_tag('a')[5].content)
    dateadded.append(line.by_tag('td')[6].content)
    cik.append(line.by_tag('td')[7].content)
```

```python
snp = pd.DataFrame({header[0]:ts, header[1]:sec, 'Sector':sector, header[3]:subi, header[5]:dateadded,
                    header[6]:cik})
snp
```

```python
def f(x):
    if "amp;" in x:
        new_x = x.split("amp;")
        y = new_x[0]+new_x[1]
        return y
    else:
        return x
```

```python
snp["GICS Sub Industry"] = snp["GICS Sub Industry"].apply(lambda x:f(x))
snp["Sector"] = snp["Sector"].apply(lambda x:f(x))
snp["Security"] = snp["Security"].apply(lambda x:f(x))
snp
```

```python
snp.to_csv("snp.csv", index=False)
```
